```python
mysqldb = pymysql.connect(user='root',password='1234',host='3.35.70.166', db='proj',charset='utf8')
curs = mysqldb.cursor()
cd = 'financedata'
########### DB에서 칼럼만 뽑기 ###################
query1 = 'show full columns from {}'.format(cd)
curs.execute(query1)
select = pd.DataFrame(curs.fetchall())
############ 칼럼 리스트 만들기 ##################
col = list()
for i in select[0]:
    col.append(i)
mysqldb.commit()
############ DB에서 데이터 가져오기 ################
curs = mysqldb.cursor()
query2 = 'select * from {}'.format(cd)
curs.execute(query2)

globals()['{}'.format(cd)] = pd.DataFrame(curs.fetchall(), columns=col)
mysqldb.commit()
mysqldb.close()
```
